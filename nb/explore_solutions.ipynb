```python
import numpy as np
import matplotlib.pyplot as plt
import os
from astropy.io import fits
import glob
%matplotlib inline
```

```python
rec_files = glob.glob("../data/reconstruction/test_*.txt")
rec_files.sort()
truth_data = {}
rec_data = {}
test_data = {}
for ii, rec_f in enumerate(rec_files[:40]):
    truth_f = rec_f.replace("data/reconstruction/test_idr", "data/truth_id")
    test_f = rec_f.replace("data/reconstruction/test_idr", "data/test_id")
    print(test_f, truth_f, rec_f)
    truth_data[ii] = np.loadtxt(truth_f)
    rec_data[ii] = np.loadtxt(rec_f)
    test_data[ii] = np.loadtxt(test_f)
```

```python
fileid=19

plt.figure(figsize=(20,4))

plt.plot(truth_data[fileid], label="Truth")
plt.plot(rec_data[fileid], label="Reconstruction")

x = np.arange(len(test_data[fileid]))
jj = test_data[fileid]>0

plt.scatter(x[jj], test_data[fileid][jj], s=30.0)
plt.xlim(500,800)
plt.legend()
```
